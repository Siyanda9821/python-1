## File handling 
- using `with` keyword
  
The below code converts a dictionary to a json and puts it into a file 

- `with` -> closes the file 
-           -> There are errors (dictionary contains functions, commas,...) 

```python
with open("./bank_dict",'w') as file:
    js.dump(bank_dict, file, indent=4)
```

```python
with open("./bank_dict", "r") as file:
    bank_acc= js.load(file)
bank_acc
```

## Summary 

- Dict -> JSON: dumps(var), dump(file)
- JSON -> Dict: loads(var), load(file)

```python
with open("./blog_post.json", "r") as file:
    blog = js.load(file)
    posts=blog['posts']
# print(type(block))
posts
```

```python
new_dict={}

for post in posts:
    title = post["title"]
    author = post["author"]
    num_comments = len(post["comments"])

    new_dict[title] = {"title":title,"author": author, "num_comments": num_comments}

new_dict
```

```python
with open("./posts_summary", "w") as file:
    js.dump(new_dict, file, indent=4)
```
